```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Male", y="CAUSE", data=data)
plt.title("Causes v/s Number of males who died in Road Accidents")
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Female", y="CAUSE", data=data)
plt.title("Causes v/s Number of females who died in Road Accidents")
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Total", y="CAUSE", data=data)
plt.title("Causes v/s Number of people who died in Road Accidents")
```

```python
sns.barplot(x="Year", y="Total", data=data)
plt.title("Year v/s Number of people who died in Road Accidents")
```

```python
data_cor=data.corr()
sns.set(rc={'figure.figsize':(10,8)})
sns.heatmap(data_cor,annot=True)
```
